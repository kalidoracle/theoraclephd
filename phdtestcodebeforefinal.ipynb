```python
import pandas as pd
from transformers import BertTokenizer, BertForTokenClassification
import torch

# Replace 'your_dataset.csv' with the actual path or URL of your CSV file
csv_file_path = '/content/output_disambiguated_concepts.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(csv_file_path)

# Load the pre-trained BERT model for Token Classification (NER)
model = BertForTokenClassification.from_pretrained("dbmdz/bert-large-cased-finetuned-conll03-english")
tokenizer = BertTokenizer.from_pretrained("dbmdz/bert-large-cased-finetuned-conll03-english")

# Function to extract concepts using BERT-based Named Entity Recognition
def extract_concepts_bert(text):
    # Tokenize the input text
    tokens = tokenizer.tokenize(tokenizer.decode(tokenizer.encode(text)))

    # Perform Named Entity Recognition using BERT model
    inputs = tokenizer.encode(tokens, return_tensors="pt")
    outputs = model(inputs).logits
    predictions = torch.argmax(outputs, dim=2)

    # Convert token predictions back to entities
    entities = []
    for token, pred_label in zip(tokens, predictions[0].numpy()):
        label = model.config.id2label[pred_label]
        if label != 'O':
            entities.append(token)

    return ' '.join(entities)

# Apply concept disambiguation to relevant columns
columns_to_disambiguate = ['text']  # Replace with your actual column names
for col in columns_to_disambiguate:
    df[col + '_disambiguated'] = df[col].apply(lambda x: extract_concepts_bert(str(x)))

# Save the DataFrame with disambiguated concepts to a new CSV file
output_csv_file_path = 'output_with_disambiguated_concepts.csv'
df.to_csv(output_csv_file_path, index=False)

# Display the first few rows of the DataFrame with disambiguated concepts
print(df.head())
```

```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained model tokenizer (vocabulary)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Load pre-trained model (weights)
model = BertModel.from_pretrained('bert-base-uncased')

# Set the model to evaluation mode
model.eval()

# Load the CSV dataset
df = pd.read_csv('/content/drive/MyDrive/imdbfortest.csv')

# Assuming the column containing the text is named 'text'
texts = df['text'].tolist()

# Function to tokenize and obtain word embeddings for a given text
# Modify the get_word_embeddings function to handle texts exceeding the maximum sequence length
def get_word_embeddings(text):
    # Tokenize the input text
    input_ids = tokenizer.encode(text, add_special_tokens=True)

    # Split the input into chunks if it exceeds the maximum sequence length
    if len(input_ids) > 512:
        chunks = [input_ids[i:i + 512] for i in range(0, len(input_ids), 512)]

        # Get word embeddings for each chunk and concatenate them
        word_embeddings = []
        for chunk in chunks:
            input_tensor = torch.tensor([chunk])
            with torch.no_grad():
                outputs = model(input_tensor)
                last_hidden_states = outputs[0]
            word_embeddings.append(last_hidden_states.squeeze(0))

        word_embeddings = torch.cat(word_embeddings, dim=0)
    else:
        # Get word embeddings for the entire text
        input_tensor = torch.tensor([input_ids])
        with torch.no_grad():
            outputs = model(input_tensor)
            last_hidden_states = outputs[0]
        word_embeddings = last_hidden_states.squeeze(0)

    return word_embeddings

# Apply the function to each text in the dataset
word_embeddings_list = [get_word_embeddings(text) for text in texts]

# Convert the list of word embeddings into a DataFrame
word_embeddings_df = pd.DataFrame(word_embeddings_list)

# Save the DataFrame containing the word embeddings
word_embeddings_df.to_csv('word_embeddings.csv', index=False)
```

```python
import spacy
from spacy.matcher import Matcher

# Load spaCy English model
nlp = spacy.load('en_core_web_sm')

# Example sentence
sentence = "Apple Inc. was founded by Steve Jobs in Cupertino."

# Process the sentence with spaCy
doc = nlp(sentence)

# Define a pattern to match "found" and "founded by" phrases
matcher = Matcher(nlp.vocab)
matcher.add("FOUNDED", [[{"LOWER": "found"}, {"POS": "VERB", "OP": "?"}, {"LOWER": "by"}]])

# Apply the matcher to the processed text
matches = matcher(doc)

# Extract relations based on matches
relations = []
for match_id, start, end in matches:
    relation = {
        "entity1": str(doc[start - 1 : start].text),
        "relation": str(doc[start : end].text),
        "entity2": str(doc[end : end + 1].text),
    }
    relations.append(relation)

# Print the extracted relations
print("Extracted Relations:")
for relation in relations:
    print(relation)
```

```python
from transformers import BertTokenizer, BertModel
import torch

def get_bert_embeddings(text):
    # Load pre-trained BERT model and tokenizer
    model_name = 'bert-base-uncased'
    tokenizer = BertTokenizer.from_pretrained(model_name)
    model = BertModel.from_pretrained(model_name)

    # Tokenize the input text
    tokens = tokenizer.tokenize(text)
    tokens = ['[CLS]'] + tokens + ['[SEP]']
    input_ids = tokenizer.convert_tokens_to_ids(tokens)
    input_ids = torch.tensor(input_ids).unsqueeze(0)  # Add batch dimension

    # Get the word embeddings
    with torch.no_grad():
        outputs = model(input_ids)
        embeddings = outputs[0]  # First element of the output tuple

    # Remove special tokens ([CLS] and [SEP]) embeddings
    embeddings = embeddings[:, 1:-1, :]  # Remove first and last tokens

    return embeddings

# Example usage
text = "Hello, how are you?"
embeddings = get_bert_embeddings(text)
print(embeddings)
```
